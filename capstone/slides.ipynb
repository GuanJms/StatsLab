```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
alpha = 1
mu = 0

# Function definition
def rho_tau(t):
    return (alpha / (np.sqrt(2 * np.pi * t ** 3))) * np.exp(-(alpha - mu * t) ** 2 / (2 * t))

def hitting_pnl(t):
    return np.power(alpha, 2) - t 

# Time values
t_values = np.linspace(0.01, 2, 400)  # Time from 0.01 to 10

# Compute rho_tau for each t
rho_values = rho_tau(t_values)

# Compute hitting_pnl for each t
pnl_values = hitting_pnl(t_values)

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(t_values, rho_values, label=r'$\rho_\tau(t) = \frac{\alpha}{\sqrt{2\pi t^3}} e^{-\frac{(\alpha-\mu t)^2}{2t}}$')
plt.plot(t_values, pnl_values, label=r'$\rho_\tau(t) = \alpha^2 - t (i.e. W_{\tau} = \alpha)$')
plt.xlabel(r'$t$')
plt.ylabel(r'$\rho_\tau(t)$')
plt.title(r'Plot of the Distribution of Hitting time $\rho_\tau(t)$')
plt.legend()
plt.grid(True)
plt.show()
```
