```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_process import ArmaProcess

# Define ARMA(2, 1) parameters
ar_params = np.array([1 , 1.2, 0])  # AR terms: y_t = 0.75*y_{t-1} - 0.25*y_{t-2}
ma_params = np.array([0., 0.00001])         # MA term: + 0.5*epsilon_{t-1}

# Create the ARMA process
arma_process = ArmaProcess(ar=ar_params, ma=ma_params)

# Simulate the process
n_points = 200
simulated_data = arma_process.generate_sample(nsample=n_points)

# Plot the simulated data
plt.figure(figsize=(12, 6))
plt.plot(simulated_data, label="ARMA(2, 1) Simulated Data")
plt.title("Simulation of ARMA(2, 1) Process")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```
