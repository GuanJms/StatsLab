```python
import numpy as np
from _lasso import lasso
from sklearn.linear_model import Lasso
import pandas as pd

def generate_dataset(n):
    x1 = np.random.normal(0, 1, n)
    x2 = np.random.normal(0, 1, n)
    x3 = np.random.normal(0, 1, n)
    epsilon = np.random.normal(0, 1, n)
    y = 3 * x1 - 17 * x2 + 5 * x3 + epsilon
    return pd.DataFrame({'x1': x1, 'x2': x2, 'x3': x3, 'y': y})

np.random.seed(0)
# Generate dataset of 100 samples
n_samples = 100
df = generate_dataset(n_samples)

# Save to CSV
csv_path = 'Simulated_Dataset.csv'
df.to_csv(csv_path, index=False)

y = df['y'].values
x1 = df['x1'].values
x2 = df['x2'].values
x3 = df['x3'].values

X = np.column_stack((x1, x2, x3))
```

```python
lmds = np.linspace(0.1, 25, 1000)
coefficients = []
beta = np.zeros(3)
for lmd in lmds:
    beta = lasso(X, y, lmd, tol=1e-12, beta=beta, intercept=False)
    coefficients.append(beta.copy())
```

```python
# Plot the coefficients
import matplotlib.pyplot as plt

coefficients = np.array(coefficients)
plt.plot(lmds, coefficients)
plt.xlabel('Lambda')
plt.ylabel('Coefficients')
plt.title('Lasso Coefficients')
plt.legend(['b1', 'b2', 'b3'])
plt.show()
```

# Verify that the coefficients are correct with Sklearn

Checking with sklearn lass

```python
# Compare with multiple lambdas and plot sum of squared differences
lmds = np.linspace(0.1, 25, 1000)
diffs = []
for lmd in lmds:
    beta = lasso(X, y, lmd, tol=1e-15, beta=beta, intercept=False)
    lasso_sklearn = Lasso(alpha=lmd, fit_intercept=False, tol=1e-15)
    lasso_sklearn.fit(X, y)
    differences = np.array(lasso_sklearn.coef_.tolist()) - beta
    diffs.append(np.sum(differences**2))

plt.plot(lmds, diffs)
plt.xlabel('Lambda')
plt.ylabel('Sum of squared differences')
plt.title('My Lasso vs. Sklearn Lasso')
plt.ylim(-1e-20, 1e-20)
```
