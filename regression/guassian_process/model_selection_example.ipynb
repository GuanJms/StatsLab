```python
from sklearn.gaussian_process.kernels import Kernel
from sklearn.gaussian_process import GaussianProcessRegressor
import numpy as np
import matplotlib.pyplot as plt

class FractionalBrownianMotionKernel(Kernel):
    def __init__(self, H):
        self.H = H

    def __call__(self, X, Y=None):
        X = np.atleast_2d(X)
        two_H = 2 * self.H
        Y = X if Y is None else np.atleast_2d(Y)
        cov = 0.5 * (np.abs(X) ** two_H + np.abs(Y.T) ** two_H - np.abs(X - Y.T) ** two_H)
        return cov

    def diag(self, X):
        return np.ones(X.shape[0])

    def is_stationary(self):
        return True
    
class OUProcessKernel(Kernel):
    def __init__(self, alpha, sigma):
        self.alpha = max(alpha, 1e-4)
        self.sigma = sigma
    
    def __call__(self, X, Y=None):
        X = np.atleast_2d(X)
        Y = X if Y is None else np.atleast_2d(Y)
        coeff = self.sigma**2 / 2 / self.alpha
        cov = coeff * (np.exp(-self.alpha * np.abs(X - Y.T)) - np.exp(-self.alpha * (X + Y.T)))
        return cov

    def diag(self, X):
        return np.ones(X.shape[0])

    def is_stationary(self):
        return True
```

```python
# Parameters
theta = 0.7    # Speed of mean reversion
mu = 100       # Long-term mean
sigma = 20     # Volatility
X0 = 100.0       # Initial value
T = 10.0       # Total time
dt = 0.01      # Time step
N = int(T / dt)  # Number of steps
t = np.linspace(0, T, N)  # Time grid

# Simulate the OU process
X = np.zeros(N)
X[0] = X0

for i in range(1, N):
    dW = np.sqrt(dt) * np.random.normal()
    X[i] = X[i-1] + theta * (mu - X[i-1]) * dt + sigma * dW
price = X.copy().reshape(1, -1)

# Plot the simulated path
plt.figure(figsize=(10, 6))
plt.plot(t, X, label="Ornstein-Uhlenbeck Process")
plt.title("Simulated Ornstein-Uhlenbeck (OU) Process")
plt.xlabel("Time")
plt.ylabel("X(t)")
plt.legend()
plt.grid()
plt.show()
```

```python
thetas = np.linspace(0, 2, 21)
sigmas = np.linspace(10, 40, 10)
likelihoods = np.zeros((len(thetas), len(sigmas)))

t = t.reshape(-1, 1)
x = price.reshape(-1,1)

# Grid search and compute log-likelihood
for i, theta in enumerate(thetas):
    for j, sigma in enumerate(sigmas):
        kernel = OUProcessKernel(alpha=theta, sigma=sigma)
        gp = GaussianProcessRegressor(kernel=kernel)        
        # Fit and compute log marginal likelihood
        gp.fit(t, x)
        likelihoods[i, j] = gp.log_marginal_likelihood()

# Plot the surface of log-likelihood
Theta, Sigma = np.meshgrid(sigmas, thetas)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(Theta, Sigma, likelihoods, cmap='viridis')

# Add labels and title
ax.set_xlabel('Sigma (Volatility)')
ax.set_ylabel('Theta (Mean Reversion)')
ax.set_zlabel('Log Marginal Likelihood')
ax.set_title('Surface of Log Marginal Likelihood')

plt.show()
```

```python
# Find the maximum likelihood and corresponding theta and sigma
max_idx = np.unravel_index(np.argmax(likelihoods), likelihoods.shape)
max_theta = thetas[max_idx[0]]
max_sigma = sigmas[max_idx[1]]
max_likelihood = likelihoods[max_idx]
```

```python
print(f"Maximum Log Marginal Likelihood: {max_likelihood}")
print(f"Optimal Theta (Mean Reversion): {max_theta}")
print(f"Optimal Sigma (Volatility): {max_sigma}")
```

```python
hs = np.linspace(0.1, 0.9, 9)
likelihoods_fbm = np.zeros((len(hs)))

# Grid search and compute log-likelihood
for i, h in enumerate(hs):
    kernel = FractionalBrownianMotionKernel(H=h)
    gp = GaussianProcessRegressor(kernel=kernel)        
    # Fit and compute log marginal likelihood
    gp.fit(t, x)
    likelihoods_fbm[i] = gp.log_marginal_likelihood()

fig = plt.figure(figsize=(10, 7))
plt.plot(hs, likelihoods_fbm)
plt.show()
```

```python
optimal_h = hs[likelihoods_fbm.argmax()]
optimal_log_marginal_likelihood = likelihoods_fbm.max()
```
